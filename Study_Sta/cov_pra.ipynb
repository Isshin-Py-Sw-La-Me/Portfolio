```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = np.array([0,0])
```

```python
cov = np.array([
  [1,0.1],
  [0.1,1]
])
```

```python
a = np.random.multivariate_normal(mean, cov, 5000)
a
```

```python
a.T
```

```python
x,y = a.T
```

```python
plt.plot(x,y,"x")
```
