```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from fastdtw import fastdtw
from scipy import spatial
```

```python
def cos_sim(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

```python
data1 = [90.0 * np.sin(i / 2) + np.random.rand() for i in range(30)]
data2 = [200.0 * np.sin(i / 2) + np.random.rand() for i in range(30)]
# data3 = [data1[i] + 100 for i in range(30)]
data3 = [90.0 * np.sin(i / 2.3) + np.random.rand() for i in range(30)]

# print(data1)
# print(data2)
# print(data3)

data_1_2_euclide = np.linalg.norm(np.array(data1) - np.array(data2))
data_1_2_cos_sim = cos_sim(np.array(data1), np.array(data2))

data_1_3_euclide = np.linalg.norm(np.array(data1) - np.array(data3))
data_1_3_cos_sim = cos_sim(np.array(data1), np.array(data3))

print(fastdtw(data1, data2)[0])
print(fastdtw(data1, data3)[0])

plt.figure(figsize=(12, 4))
plt.title(
    "Euclid: D(data1, data3) = {} > D(data1, data2) = {}\nCos_sim: D(data1, data3) = {} < D(data1, data2) = {}".format(
        data_1_3_euclide, data_1_2_euclide, data_1_3_cos_sim, data_1_2_cos_sim
    )
)
plt.plot(data1, label="data1")
plt.plot(data2, label="data2")
plt.plot(data3, label="data3")
plt.legend()
```

```python
hoge = np.array([1, 2])
foo = np.array([2, 3])
baa = np.array([1, 1])

print(fastdtw(hoge, foo)[0])
print(fastdtw(hoge, baa)[0])

# print(fastdtw(hoge, foo))
# print(fastdtw(hoge, baa))

plt.title(
    "Euclid: D(data1, data3) = {} > D(data1, data2) = {}\nCos_sim: D(data1, data3) = {} < D(data1, data2) = {}".format(
        np.linalg.norm(hoge - baa),
        np.linalg.norm(hoge - foo),
        cos_sim(hoge, baa),
        cos_sim(hoge, foo),
    )
)
plt.plot(hoge, label="data1")
plt.plot(foo, label="data2")
plt.plot(baa, label="data3")
plt.legend()
# print("Cos_similarity: {}".format(cos_sim(hoge, foo)))
# print("Euclid_distance: {}".format(np.linalg.norm(hoge - foo)))
```
