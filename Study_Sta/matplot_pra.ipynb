```python
xmin, xmax = -np.pi,np.pi
x = np.arange(xmin,xmax,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.hlines([-1, 1], xmin, xmax, linestyles="dashed")  # y=-1, 1
plt.title(r"$\sin(x)$ and $\cos(x)$")
plt.xlim(xmin, xmax)
plt.ylim(-1.3, 1.3)
```

```python
#plt.subplot()
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.xlim(xmin,xmax)
plt.ylim(-1.3,1.3)
```

```python
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.xlim(xmin,xmax)
plt.ylim(-1.3,1.3)
```

```python
x = np.arange(-3,3,0.01)
y_sin = np.sin(x)
y_cos = np.cos(x)

#1. Figure
fig = plt.figure()
print("type(fig): {}".format(type(fig)))
```

```python
#2. Axes
#add_subplot(,,)
# add_subplot(111) = add_subplot(1,1,1)
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

#3. 
ax1.plot(x,y_sin)
ax2.plot(x,y_cos)

#4. y
ax1.set_ylim(-1.3, 1.3)
ax2.set_ylim(-1.3, 1.3)

ax1.set_title("$\sin x$")
ax2.set_title("$\cos x$")

ax1.set_xlabel("x")
ax2.set_xlabel("x")

fig.tight_layout()
```
