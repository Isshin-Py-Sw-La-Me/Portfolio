```python
#格子点の作成
# 座標平面上にある点でx,y座標ともに整数である点を格子点
# 格子点の数 = len(x) * len(y)
import numpy as np
x = [0, 1, 2, 3]
y = [0, 1, 2, 3]
X,Y = np.meshgrid(x,y)
X,Y
```

```python
import numpy as np
import matplotlib.pyplot as plt

# z = 2(x + y)
def f(x, y):
    return 2 * (x + y)

x = [-1, 0, 1, 2, 3]
y = [-1, 0, 1, 2, 3]

X, Y = np.meshgrid(x, y)
X,Y
```

```python
fig, ax = plt.subplots()
cont = ax.contour(X, Y, f(X, Y), levels=[0, 2, 4]) # ここだけ説明。
cont.clabel(fmt='%1.1f', fontsize=14)
plt.show()
```
