```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
from matplotlib import pyplot as plt
plt.rcParams['font.family'] = 'Hiragino Maru Gothic Pro'

df = pd.read_csv("./datasets/fixed_battery_log_2.csv")
df["date"] = pd.to_datetime(df["date"])
users = df["User"].unique()
users
```

```python
for g in range(len(users)):
    x_range = np.arange(0,86400,3600)
    first = 1
    last = 7
    color = ["red","green","blue","yellow"]
    user = users[g]
    for i in range(4):
        plt.subplot(5,1,i+1)
        plt.xticks(x_range)
        plt.ylim(0,10)
        plt.hist(df.loc[(df["User"] == user) & 
                            (df["state"] == "ON") & 
                            (df["date"] > dt.datetime(2021,12,first)) & 
                            (df["date"] < dt.datetime(2021,12,last))]["time_to_num"],
        #              range=(0,86400),
                     bins=24,
                     ec="black",
                     color = color[i],
    #                  alpha=0.3
                )
        if i == 3:
            first = first+7
            last = 31
        else:
            first = first+7
            last = last+7
    
    plt.subplot(5,1,5)
    plt.xticks(x_range)
    plt.ylim(0,10)
    first = 1
    last = 7
    for i in range(4):
        plt.hist(df.loc[(df["User"] == user) & 
                            (df["state"] == "ON") & 
                            (df["date"] > dt.datetime(2021,12,first)) & 
                            (df["date"] < dt.datetime(2021,12,last))]["time_to_num"],
        #              range=(0,86400),
                     bins=24,
                     ec="black",
                     color = color[i],
                     alpha=0.7
                )
        if i == 3:
            first = first+7
            last = 31
        else:
            first = first+7
            last = last+7
    
    plt.suptitle(user)
#     plt.savefig("bs_2_results/T_{}_simple.jpg".format(user))
    plt.clf()
```

```python
plt.figure(figsize=(80,30))
for h in range(len(users)):
    plt.subplot(2,4,h+1)
    user = users[h]
    img = plt.imread("./bs_2_results/T_{}_simple.jpg".format(user))
    plt.imshow(img)
# plt.savefig("bs_2_results/T_all_simple.jpg")
```
